### Handling missing values 

```python
#Calculating the mean of the Age column 
mean = titanic_data.age.mean()
```

```python
#Filling the missing values in the Age column with the mean
titanic_data['age'].fillna(value = mean , inplace=True)
```

```python
# Dropping the Column Deck from the dataset 
titanic_data.drop(columns='deck' , axis =1 , inplace=True)
```

```python
#Displaying the data 
titanic_data
```

```python
#filling the null vallues in embarked and embarked_town with their nearest values
titanic_data['embark_town'].fillna(method='ffill',inplace=True)
titanic_data['embarked'].fillna(method='ffill',inplace=True)
```

```python
#Confirming to see if there is no more missing values 
titanic_data.isnull().sum()
```

## Feature Engineering

```python
# Adding Two columns to form the Family Size column
titanic_data['Family_Size'] = 0
titanic_data['Family_Size'] = titanic_data['parch']+titanic_data['sibsp']
titanic_data
```
